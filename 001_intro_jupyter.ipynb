## Jupyter Notebook

O Jupyter é uma aplicação de código aberto que permite criar e compartilhar documentos que contenham um código iterativo, equações, formulas matemáticas, imagens e gráficos.

Documentação:  https://docs.jupyter.org/en/latest/index.html <br>
Cheatsheet: http://datacamp-community-prod.s3.amazonaws.com/21fdc814-3f08-4aa9-90fa-247eedefd655 <br>
28 Tips: https://www.dataquest.io/blog/jupyter-notebook-tips-tricks-shortcuts/ <br>

As células acima possuem texto em formato markdown ([saiba + sobre markdown](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)) que são renderizadas assim que a célula é executada. Já a célula abaixo possui código e podemos ver uma clara diferença.

```python
from matplotlib import pyplot as plt
import numpy as np

# gerando 100 pontos aleatórios
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots(figsize=(9,6))

# fazendo o gráfico
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Gráfico feito com python no Jupyter Notebook")
plt.show()
```

#### Botões:
___

1. save and checkpoint: salvar o notebook e criar um *checkpoint* das saídas.
2. insert cell below: inserir uma nova célula vazia abaixo da atual.
3. cut selected cells: recortar as células selecionadas.
4. copy selected cells: copiar células selecionadas.
5. paste cells below: colar as células recortadas ou copidas abaixo da atual.
6. move selected cells up: movimentar as células selecionadas para cima.
7. move selected cells down: movimentas as células selecionadas para baixo.
8. run cells, select below: executar a célula atual e selecionar a célula imediatamente abaixo.
9. interrupt the kernel: interromper a execução do *notebook* por completo.
10. restart the kernel (with dialog): reiniciar o *notebook*.
11. restart the kernel, then re-run the whole notebook: reiniciar o *notebook* e re-executar todas as células.
12. cell type: escolher o tipo da célula selecionada.
